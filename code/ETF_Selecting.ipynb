```python
#爬取ETF Category等信息
#整理ETF的Category信息
import requests
from lxml import html
import re
import pandas as pd 
ETF_Information={key:{} for key in set7}
# 目标网页的 URL
for ticker in set7:
    url = 'https://stockanalysis.com/etf/'+ticker.lower()+'/'

# 发送 HTTP GET 请求，获取网页内容
    response = requests.get(url)

# 确保请求成功
    if response.status_code == 200:
    # 使用 lxml 解析网页内容
        tree = html.fromstring(response.content)

    # 提取特定的 XPath 元素
        xpath_expression = '//*[@data-test="overview-profile-description"]/..'
        elements = tree.xpath(xpath_expression)

    # 如果找到了元素，打印出来
        if elements:
            for element in elements:
                text=element.text_content().strip()
        else:
            print("没有找到匹配的元素。")
    else:
        print(f"请求失败，状态码：{response.status_code}")
    keywords = {'Asset Class': 'asset_class', 'Category': 'category','Index Tracked': 'index_tracked'}
    for keyword,key in keywords.items():
        pattern = re.compile(r'{} (\S+)'.format(re.escape(keyword)), re.IGNORECASE)
        match = pattern.search(text)
        if match:
            ETF_Information[ticker][key] = match.group(1)
    series=pd.Series(ETF_Information[ticker])
    series.to_csv('/Users/insnap/Desktop/BTC_data_analyzation/code/ETF_Information'+ticker+'.csv')
    print(ticker)
print(ETF_Information)
```
